Define `reflect(attempt, is_good, critique, max_retries=2)`:
- call `attempt()`; if `is_good(result)` is False, call `critique(result)` and retry — up to `max_retries` extra times.
- return a dict: `{'result', 'success', 'retries_used', 'critiques'}`.

Graded by injecting scripted `attempt`/`is_good` — measures that you improve and respect the budget.

```python
def reflect(attempt, is_good, critique, max_retries=2):
    critiques, used = [], 0
    result = attempt()
    while not is_good(result) and used < max_retries:
        critiques.append(critique(result))
        used += 1
        result = attempt()
    return {'result': result, 'success': is_good(result),
            'retries_used': used, 'critiques': critiques}
```

```python
# scripted check: succeeds on the 3rd attempt
box = {'n': 0}
def attempt():
    box['n'] += 1; return box['n']
rec = reflect(attempt, lambda r: r >= 3, lambda r: f'too small {r}', max_retries=5)
print(rec['success'], rec['retries_used'])  # True 2
```
