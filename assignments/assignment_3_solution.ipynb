# Assignment 3 — retrieval into context
Implements: **assignment_3 (Module 3)**

Define:
- `retrieve(query_vec, doc_vecs, k=3)` → list of doc indices, best first.
- `build_context(query_vec, doc_vecs, docs, k=3)` → a string containing the top-k docs (one per line).

Use cosine similarity. Graded with a **fixed embedding fixture** so scoring is deterministic.

```python
import os, sys
sys.path.insert(0, os.path.abspath(os.path.join('..', 'src')))
```

```python
import numpy as np
```

```python
def cosine(a, b):
    a, b = np.asarray(a, float), np.asarray(b, float)
    d = np.linalg.norm(a) * np.linalg.norm(b)
    return 0.0 if d == 0 else float(np.dot(a, b) / d)

def retrieve(query_vec, doc_vecs, k=3):
    scored = sorted(range(len(doc_vecs)),
                    key=lambda i: cosine(query_vec, doc_vecs[i]), reverse=True)
    return scored[:k]
```

```python
def build_context(query_vec, doc_vecs, docs, k=3):
    idx = retrieve(query_vec, doc_vecs, k)
    return '\n'.join(docs[i] for i in idx)
```

```python
import json
FIX = os.path.abspath(os.path.join('..', 'fixtures', 'embeddings'))
vecs = np.load(os.path.join(FIX, 'doc_vectors.npy'))
docs = json.load(open(os.path.join(FIX, 'docs.json')))
print(build_context(vecs[0], vecs, docs, k=2))  # expect the two France lines
```
