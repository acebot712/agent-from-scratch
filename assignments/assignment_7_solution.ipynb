# Assignment 7 — caps & logging
Implements: **assignment_7 (Module 7)** · fully deterministic

Define:
- `cap(steps, cost, max_steps, max_cost)` → a halt-reason string, or `None` if within both caps. (Check steps first.)
- `log_record(step, action, cost, cumulative)` → a flat dict with keys `step, action, cost_usd, cumulative_cost_usd, ok`.

Minimal scaffold — you write the bodies.

```python
import os, sys
sys.path.insert(0, os.path.abspath(os.path.join('..', 'src')))
```

```python
def cap(steps, cost, max_steps, max_cost):
    if steps >= max_steps:
        return f'max_steps ({max_steps}) reached'
    if cost >= max_cost:
        return f'max_cost (${max_cost}) reached'
    return None

def log_record(step, action, cost, cumulative):
    return {'step': step, 'action': action, 'cost_usd': round(cost, 6),
            'cumulative_cost_usd': round(cumulative, 6), 'ok': True}
```

```python
print(cap(5, 0.1, 5, 1.0))     # max_steps (5) reached
print(cap(2, 1.0, 5, 1.0))     # max_cost ($1.0) reached
print(cap(2, 0.1, 5, 1.0))     # None
print(log_record(1, 'llm_call', 0.01, 0.01))
```
