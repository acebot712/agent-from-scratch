# Assignment 2 — tools, registry, dispatch
Implements: **assignment_2 (Module 2)**

Define:
- `build_registry()` → a `ToolRegistry` with **three** tools named `add`, `upper`, `length`.
- `handle(text, registry)` → parse a tool request from `text`, dispatch it, and return the `ToolResult`.

Graded on correct tool selection and graceful unknown-tool handling — no LLM needed.

```python
import os, sys
sys.path.insert(0, os.path.abspath(os.path.join('..', 'src')))
```

```python
from agent.tools import Tool, ToolRegistry, parse_tool_call, dispatch
```

```python
def build_registry():
    reg = ToolRegistry()
    reg.register(Tool('add', 'add a and b', lambda a, b: a + b,
        {'type':'object','properties':{'a':{'type':'number'},'b':{'type':'number'}}}))
    reg.register(Tool('upper', 'uppercase s', lambda s: s.upper(),
        {'type':'object','properties':{'s':{'type':'string'}}}))
    reg.register(Tool('length', 'length of s', lambda s: len(s),
        {'type':'object','properties':{'s':{'type':'string'}}}))
    return reg
```

```python
def handle(text, registry):
    call = parse_tool_call(text)
    return dispatch(registry, call)
```

```python
reg = build_registry()
print(handle('TOOL: add\nARGS: {"a": 2, "b": 3}', reg).output)   # 5
print(handle('TOOL: ghost\nARGS: {}', reg).ok)                       # False
```
