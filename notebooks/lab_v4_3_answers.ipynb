```python
from agent import Tool, ToolRegistry, run_react, complete

def calc(x):
    """Evaluate a simple arithmetic string, e.g. '128*47'."""
    return str(eval(x, {'__builtins__': {}}, {}))

reg = ToolRegistry()
reg.register(Tool(name='calc', description='evaluate arithmetic; arg is the expression',
                  run=calc, parameters={'type':'object','properties':{'x':{'type':'string'}}}))
```

```python
TASK = 'What is 128 * 47, then add 19? Reason carefully.'

# Naive: one call, no tools.
naive = complete([{'role':'user','content': TASK}]).text
print('NAIVE:\n', naive)
```

```python
react = run_react(TASK, reg, max_steps=6)
print('REACT:\n', react)
```

```python
expected = 128 * 47 + 19
print('expected:', expected, '| react correct:', str(expected) in react)
```
