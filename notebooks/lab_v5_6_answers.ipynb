```python
from agent import Coordinator, Worker, DelegationCap, synthesize_outputs

researcher = Worker('research', 'You research facts. Reply with 2 bullet facts.')
writer = Worker('write', 'You write a one-sentence summary from the input.')
```

```python
reviewer = Worker('review', 'You critique the summary in one short sentence.')
```

```python
coord = Coordinator(
    workers={'research': researcher, 'write': writer, 'review': reviewer},
    cap=DelegationCap(max_depth=3),
    decompose=lambda task: [('research', task), ('write', task), ('review', task)],
)
```

```python
# Run it (needs a key). Offline? Skip this and run the stub cell below.
result = coord.run('the benefits of cycling to work')
print(result['synthesis']['combined'])
```
