```python
from agent import Tool, ToolRegistry, ToolAgent

# A fake weather service so the lab is reproducible.
_WEATHER = {'Paris': '18C, cloudy', 'Cairo': '34C, sunny'}
```

```python
def get_weather(city):
    return _WEATHER.get(city, f'no data for {city}')
```

```python
registry = ToolRegistry()
registry.register(Tool(
    name='get_weather', description='current weather for a city; arg: city',
    run=get_weather,
    parameters={'type': 'object', 'properties': {'city': {'type': 'string'}}},
))
print(registry.describe())
```

```python
agent = ToolAgent(registry, max_steps=5)
print(agent.run('What is the weather in Paris right now?'))
```

```python
# Did it actually call the tool? Look for an Observation turn:
for m in agent.history:
    print(m['role'], '·', m['content'][:80])
```
