```python
from agent.memory import EpisodicMemory
PATH = 'lab_memory.json'
```

```python
mem = EpisodicMemory()
mem.record('user', 'My favorite color is teal.')
mem.save(PATH)
```

```python
reloaded = EpisodicMemory.load(PATH)
for e in reloaded.episodes:
    print(e.role, '·', e.content)
```
