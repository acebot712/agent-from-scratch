```python
# Sample model responses — some call a tool, some answer directly.
samples = [
    'Thought: I need the weather.\nACTION: get_weather("Paris")',
    'ACTION: add(2, 3)',
    'The answer is 5. FINAL ANSWER: 5',
]
```

```python
import re
def parse_intent(text):
    if 'FINAL ANSWER:' in text:
        return {'kind': 'final', 'value': text.split('FINAL ANSWER:', 1)[1].strip()}
    m = re.search(r'ACTION:\s*([A-Za-z0-9_]+)\((.*)\)', text)
    if m:
        return {'kind': 'action', 'name': m.group(1), 'raw_args': m.group(2).strip()}
    return {'kind': 'unknown'}
```

```python
for s in samples:
    print(parse_intent(s))
```
