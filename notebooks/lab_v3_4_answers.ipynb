**Learning objective:** compute cosine similarity and return the top-k most similar docs. Runs offline on the shipped embedding fixture.

```python
import os, sys
sys.path.insert(0, os.path.abspath(os.path.join('..', 'src')))
# (labs import from src/agent so they track the framework you build)
```

```python
import json, numpy as np
FIX = os.path.abspath(os.path.join('..', 'fixtures', 'embeddings'))
vecs = np.load(os.path.join(FIX, 'doc_vectors.npy'))
docs = json.load(open(os.path.join(FIX, 'docs.json')))
print(vecs.shape)
for d in docs: print('-', d)
```

```python
def cosine(a, b):
    a, b = np.asarray(a, float), np.asarray(b, float)
    denom = np.linalg.norm(a) * np.linalg.norm(b)
    return 0.0 if denom == 0 else float(np.dot(a, b) / denom)
```

```python
def top_k(query_vec, doc_vecs, k=2):
    scored = [(i, cosine(query_vec, v)) for i, v in enumerate(doc_vecs)]
    scored.sort(key=lambda t: t[1], reverse=True)
    return scored[:k]
```

```python
query = vecs[0]  # 'capital of France' — geography cluster
for i, s in top_k(query, vecs, k=2):
    print(round(s, 3), docs[i])
```
